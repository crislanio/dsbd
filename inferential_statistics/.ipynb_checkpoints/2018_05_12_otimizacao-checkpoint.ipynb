```r
# reading the data, separated by space:

df <- read.table('./data/2018_05_12_youtube.txt', sep = ' ')
```

```r
# structure of the data.frame:

str(df)
```

```r
# how many unique channels?

unique(df$CANAL)
```

```r
# subsetting by channel with cols DIAS and INSCRITOS,
# replace INSCRITOS with its cumulative sum
# divide it by 100,000:
inventonahora <- df[which(df$CANAL == 'inventonahora'), c('DIAS', 'INSCRITOS')]
inventonahora$INSCRITOS <- cumsum(inventonahora$INSCRITOS)
inventonahora$INSCRITOS <- inventonahora$INSCRITOS / 100000

vocesabia <- df[which(df$CANAL == 'vocesabia'), c('DIAS', 'INSCRITOS')]
vocesabia$INSCRITOS <- cumsum(vocesabia$INSCRITOS)
vocesabia$INSCRITOS <- vocesabia$INSCRITOS / 100000
rownames(vocesabia) <- NULL
```

```r
# plotting vocesabia:

plot(x = vocesabia$DIAS,
     y = vocesabia$INSCRITOS,
     xlab = 'Days since channel creation',
     ylab = 'Cumulative sum of subscribers (divided by 100,000)',
     main = 'Vocêsabia subscribers growth')
```

```r
# plotting inventonahora:

plot(x = inventonahora$DIAS,
     y = inventonahora$INSCRITOS,
     xlab = 'Days since channel creation',
     ylab = 'Cumulative sum of subscribers (divided by 100,000)',
     main = 'Inventonahora subscribers growth')
```

# inventonahora fitting

```r
# "understanding" the data:
p_df <- data.frame(DIAS = seq(1, max(inventonahora$DIAS) + 365), INSCRITOS = NA)
p_df$INSCRITOS <- 28 / (1 + exp(-0.008 * (p_df$DIAS - 700)))

# "good" starting values:
L <- 24; b1 <- -0.007; b0 <- -600
L <- 29; b1 <- -0.008; b0 <- -700

plot(x = p_df$DIAS,
     y = p_df$INSCRITOS,
     col = 'green',
     main = '"Understanding" the data...')

lines(x = inventonahora$DIAS,
      y = inventonahora$INSCRITOS,
      col = 'red',
      lwd = 10)

legend('topleft', c('Data', 'Predicted'), lty = c(1, 1), lwd = c(5, 5), col = c('red', 'green'))
```

```r
# defining the distribution and loss function:

fn <- function(par, y, x) {
    mu <- par[3] / ( 1 + exp( par[2] * (x + par[1]) ) )
    SQ <- sum((y - mu) ^ 2)
    return(SQ)
}
```

```r
# fitting the data with optim function:

fit <- optim(par = c(-600, -0.007, 24),
             fn = fn,
             y = inventonahora$INSCRITOS,
             x = inventonahora$DIAS)
fit
```

```r
# making a prediction data.frame:

predicted_df <- data.frame(DIAS = seq(1, max(inventonahora$DIAS) + 365), INSCRITOS = NA)
predicted_df$INSCRITOS <- fit$par[3] / ( 1 + exp(fit$par[2] * (predicted_df$DIAS + fit$par[1]) ) )
```

```r
# plotting the data and the prediction:

plot(predicted_df$DIAS,
     predicted_df$INSCRITOS,
     col = 'green',
     xlab = 'Days since channel creation',
     ylab = 'Cumulative sum of subscribers (divided by 100,000)',
     main = 'inventonahora: Data vs Predicted subscribers')

lines(x = inventonahora$DIAS,
      y = inventonahora$INSCRITOS,
      col = 'red',
      lwd = 10)

legend(x = 'topleft',
       legend = c('Data', 'Prediction'),
       lty = c(1, 1),
       lwd = c(5, 5),
       col = c('red', 'green'))
```
